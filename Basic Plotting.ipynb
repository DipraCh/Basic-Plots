```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Plot Y=X
num_samples =10
x = np.linspace(2,10,num_samples)
y = x 
plt.plot(x,y)
```

```python
#Plot Bivariate Data
num_samples =10
x = np.linspace(-10,10,num_samples)
y = x **2 + x +3
plt.plot(x,y)
```

```python
#Random Plotting
num_samples =10
x = np.random.randint(0,10,20)
y = np.random.randint(0,10,20)
plt.plot(x,y)
```

```python
#Random Plotting
num_samples =10
x = np.random.randint(0,10,20)
y = np.random.randint(0,10,20)
plt.figure(figsize=(10,5))
plt.plot(x,y)
```

```python
#Plot Y=Sin(X)
num_samples =200
x = np.linspace(2,10,num_samples)
y = np.sin(x)
plt.plot(x,y)
```

```python
#Plot Multiple Lines with Labels
num_samples = 5
num_lines = 5
for i in range(0,num_lines):
    x = np.linspace(2,10,num_samples)
    y = (i/num_lines)*x +1
    plt.plot(x,y,label="slope_"+str(i))
    plt.legend()
```

```python
#Plot Circle 
num_samples =1000
x = np.linspace(-1,1,num_samples)
y = np.sqrt(1-x**2)
plt.plot(x,y,color='b')
plt.plot(x,-y,color='b')
```

```python
#Scatter Plot
num_samples =10
x = np.random.uniform(1,5,(num_samples,2))
print(x.shape)

plt.scatter(x[:,0],x[:,1],c="r",label="First Data",marker="s")

num_samples =10
x = np.random.uniform(15,20,(num_samples,2))
print(x.shape)

plt.scatter(x[:,0],x[:,1],c="g",label="Second Data",marker="s")
plt.legend()
```

```python
# scatter markers
# scatter size

# scatter plot
num_samples=50
data=np.random.uniform(1,3.5,(num_samples,2))
print(data.shape)

alphas = np.linspace(0.1, 1, num_samples)
rgba_colors = np.zeros((num_samples,4))
# for red the first column needs to be one
rgba_colors[:,0] = 1.0
# the fourth column needs to be your alphas
rgba_colors[:, 3] = alphas

plt.figure(figsize=(10,10))
plt.scatter(data[:,0],data[:,1],c=rgba_colors,s=40,marker="o",label="women")

num_samples=150
data=np.random.uniform(2,5,(num_samples,2))
print(data.shape)
markersizes=np.linspace(10,80,num_samples)
plt.scatter(data[:,0],data[:,1],color="g",s=markersizes,marker='s',label="men")

plt.title("Analysis of pro basketball performance w.r.t height, gender, and age")
plt.xlabel("average score")
plt.ylabel("height")
_=plt.legend()
```

# Subplot

```python
import math 

num_samples=150

# subplots example
NUM_PLOTS=8
num_columns=4
num_rows=NUM_PLOTS//4
fig,axes=plt.subplots(num_rows,num_columns,squeeze=False,figsize=(4*num_columns,4*num_rows))
for i in range(0,NUM_PLOTS):
    myax=axes[math.floor(i/num_columns)][i%num_columns]
    myax.set_title("y=x^"+str(i),fontweight="bold",size="12")
    x=np.linspace(-5,5,num_samples)
    y=x**i
    myax.plot(x,y,color="g")
```
